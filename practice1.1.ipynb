## Задача 1.1. Меры расстояния и их применимость

**Условие.**
Перечислите и кратко охарактеризуйте следующие меры расстояния между объектами:

1. Евклидово расстояние.
2. Манхэттенское расстояние.
3. Минковского расстояние (Lp‑норма).
4. Чебышёвское расстояние (L∞).
5. Хэммингово расстояние.
6. Косинусное расстояние (на основе косинусного сходства).
7. Джаккардово расстояние.

Для каждой меры укажите:
- тип данных (числовые / бинарные / категориальные / векторные), для которых она корректно применяется;
- особенности (чувствительность к масштабу, выбросам, разреженности).


1. Евклидово расстояние.  $$d_{Eu}(x,y)=\sqrt{\sum_{i=1}^N (x_i-y_i)^2}$$
2. Манхэттенское расстояние. Сумма расстояний по проекциям на координатные оси  $$d_{Mh}(x,y)=\sum_{i=1}^N |x_i-y_i|$$
3. Минковского расстояние (Lp‑норма). Обобщение для Евклидова и Манхэттеновского расстояний  $$d_{p}(x,y)= \left(\sum_{i=1}^N |x_i-y_i|^p\right)^{1 \over p}$$
4. Чебышёвское расстояние. Берется максимум от модуля разницы координат в любом направлении  $$d_{∞}(x,y)=\max_{i=1,...,N} |x_i-y_i|$$
5. Хэммингово расстояние. 
$$d_h(x,y)= count_{i=1}^N(x_i \ne y_i)$$
6. Косинусное расстояние (на основе косинусного сходства). $$d_{\text{cos}}(x,y)=1-\frac{\sum_{i=1}^N x_i y_i}{\sqrt{\sum_{i=1}^N x_i^2}\sqrt{\sum_{i=1}^N y_i^2}}$$
7. Джаккардово расстояние. отношение размера пересечения множеств к размеру их объединения  $$d_{\text{Jaccard}}(A,B)=1 - \frac{A \cap B}{A \cup B} $$

Для каждой меры укажите:
- тип данных (числовые / бинарные / категориальные / векторные), для которых она корректно применяется;
- особенности (чувствительность к масштабу, выбросам, разреженности).

1. Евклидово расстояние

    Тип данных: Непрерывные числовые (количественные) признаки.
    Особенности: 
    • Масштаб: Высокая чувствительность. Признаки с разными единицами измерения или диапазонами искажают результат, поэтому обязательна нормализация или стандартизация.
    • Выбросы: Крайне уязвимо. Возведение разностей в квадрат многократно усиливает вклад аномальных значений.
    • Разреженность: Работает неэффективно. В пространствах с большим количеством нулей теряет различимость объектов.
    • Прочее: Корректно только при допущении о некоррелированности признаков и их равной значимости (без весовых коэффициентов).

2. Манхэттенское расстояние

    Тип данных: Числовые (количественные).
    Особенности: 
    • Масштаб: Чувствительно, также требует приведения признаков к единому диапазону.
    • Выбросы: Устойчивее евклидового, так как использует модуль разности, а не квадрат.
    • Разреженность: Сохраняет информативность в высокоразмерных разреженных пространствах, часто предпочтительнее евклидовой метрики при d >> n.
    • Прочее: Геометрически соответствует перемещению по осям сетки (L1-норма).

3. Расстояние Минковского

    Тип данных: Числовые.
    Особенности: 
    • Масштаб: Зависит от масштаба, нормализация данных обязательна.
    • Выбросы: Управляемая чувствительность через параметр p. Рост p усиливает влияние крупных отклонений, снижение p делает метрику робастнее.
    • Разреженность: Поведение наследует свойства Lp-нормы; при p=1 или p=2 характеристики аналогичны Манхэттену и Евклиду соответственно.
    • Прочее: При p → ∞ асимптотически переходит в расстояние Чебышёва. Требует валидации для подбора p, хотя на практике чаще фиксируют 1 или 2.

4. Расстояние Чебышёва

    Тип данных: Числовые.
    Особенности: 
    • Масштаб: Чувствительно, но только по той координате, где разница максимальна.
    • Выбросы: Индифферентно к аномалиям в остальных измерениях, но критично зависит от выброса именно в координате с максимальным отклонением.
    • Разреженность: Не оптимально для разреженных данных, так как полностью игнорирует согласованность по другим признакам.
    • Прочее: Определяется исключительно по наибольшей абсолютной разнице. Востребовано в задачах, где лимитирующим фактором является худшее отклонение (логистика, управление движением, допуски в инженерии).

5. Расстояние Хэмминга

    Тип данных: Бинарные данные, бинарные векторы, строки одинаковой длины.
    Особенности: 
    • Масштаб: Не применимо (данные дискретны: 0/1).
    • Выбросы: Понятие выброса заменяется на несовпадение позиций. Все ошибки вносят равный аддитивный вклад.
    • Разреженность: Эффективно, так как подсчитывает только число различий, игнорируя абсолютные величины.
    • Прочее: Идеально для сравнения флагов, хешей, генетических кодов или исправления ошибок в передачах данных.

6. Косинусное расстояние

    Тип данных: Числовые векторы (преимущественно высокоразмерные).
    Особенности: 
    • Масштаб: Полностью инвариантно. Зависит только от угла между векторами, а не от их абсолютной длины.
    • Выбросы: Умеренно чувствительно, но влияние аномалий сглаживается нормировкой на длину вектора.
    • Разреженность: Отлично адаптировано. Широко применяется в NLP и рекомендательных системах, где данные естественно разрежены (TF-IDF, эмбеддинги).
    • Прочее: Измеряет семантическую или тематическую схожесть направлений, а не абсолютное совпадение значений.

7. Расстояние Жаккара

    Тип данных: Множества, бинарные векторы, наборы тегов/токенов.
    Особенности: 
    • Масштаб: Не зависит от масштаба, реагирует только на факт присутствия/отсутствия элемента.
    • Выбросы: Устойчиво, так как учитывает лишь отношение пересечения к объединению множеств.
    • Разреженность: Оптимально для разреженных бинарных данных. Игнорирует совместное отсутствие признаков (нуль-нуль пары), что важно при сравнении больших словарей.
    • Прочее: Нормировано на отрезок [0, 1], где 0 означает полное совпадение, а 1 — полное непересечение. Часто используется для дедупликации текстов и анализа корзин покупок.

Используя Хэммингово расстояние пытаемся распределить объекты А и В с $(Z_1,Z_2) = (1,0)$ и $(0,1)$ по классам. Получаем:  
А: $d_{h_{z^{(I)}}} = 1$  $d_{h_{z^{(II)}}} = 1$  
B: $d_{h_{z^{(I)}}} = 1$  $d_{h_{z^{(II)}}} = 1$  
Распределить объекты А и В по классам используя исключительно хэммингово расстояние до прототипов не представляется возможным. 

```python
import numpy as np
import pandas as pd

def g(df: pd.DataFrame):
    return df['Z1'].map(lambda x: "I" if x >= 1 else "II")

# Все возможные векторы (Z1, Z2)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Классы: 1 = I, 0 = II
y = np.array([0, 0, 1, 1])

df_train = pd.DataFrame(X, columns=["Z1", "Z2"])
df_train["class_label"] = y
df_train["class_name"] = df_train["class_label"].map({0: "II", 1: "I"})
df_train["classified"] = g(df_train)
df_train
```
