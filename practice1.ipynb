#### Система проверки банковских транзакций
 - Множество объектов: профили отправителя и получателя (их истории транзакций / статус), транзакция - сумма и дата
 - Множество классов: мошенническая, не мошенническая 
 - Тип задачи: бинарная классификация
 - Данные: Дата, сумма, тип, геопозиция, статистика транзакций пользователей, их статусы и устройства.
 - Знания: Паттерны поведения злоумышленников, эксперные оценки операторов
 - Гипотеза: Предполагается, что в поведении мошенников есть паттерны
 - Закономерность: Статистическая закономерность в имеющихся данных о пользователях и транзакциях, которая позволит по признаковым показаням транзакции позволят определить вероятность того, что она мошенническая 
 - Приоритетный тип ошибки: Ложно-отрицательные

#### Классификация входящих писем как «спам / не спам»
 - Множество объектов: Электронные письма
 - Множество классов: Спам / не спам
 - Тип задачи: бинарная классификация
 - Данные: Заголовок, отправитель, список получателей, тема, текст сообщения, вложение, ссылки, список "отмечено как спам"
 - Знания: База отправителей - спамеров,  база спам сообщений и вредоносных ссылок, база официальных интернет-магазинов / банков и т.п.
 - Гипотеза: В структуре отправляемых спам-сообщений, в тексте, ссылках и прикрепляемых файлах есть часто повторяющиеся паттерны
 - Закономерность: В спам сообщенях часто фигурируют одни и те же темы, например: перейдите по ссылке, откройте документ, и они позволяют определить спам.
 - Приоритетный тип ошибки: Важно как отобразить все нормальные сообщения и отсеять спам.

#### Распознавание дефектных изделий по фотографиям на конвейере
 - Множество объектов: Фото
 - Множество классов: Норма и брак
 - Тип задачи: Выявление аномалий 
 - Данные: Нынешнее фото, фото нормальных деталей, фото бракованного изделия
 - Знания: Экспертные сведения о возможных критических браках и их локализации по фото
 - Гипотеза: Бракованное изделие будет визуально отличаться по фото
 - Закономерность: Внешний вид критических областей совпадает у нормальных деталей, и отличается у бракованных
 - Приоритетный тип ошибки: Нахождение бракованных деталей

## Задача 2. Формализация задачи классификации в терминах $X, Y, f$

**Условие.**
Задача: автоматическое определение языка короткого текстового сообщения (русский, английский, китайский).

### Требуется:
1. Определить:
   - пространство объектов $X$ (что является наблюдаемыми данными);
   - множество меток классов $Y$;
   - вид функции распознавания $f: X \to Y$.
2. Сформулировать условие «правильного» распознавания в виде:
   
   $\forall (x,y)\in \mathcal{D}:\ f(x) \approx y$,
   
   где $\mathcal{D}$ — распределение пар «текст–язык».
3. Обсудить, какие априорные знания о языках можно использовать.


### Ваше решение задачи 2

_Опишите здесь X, Y, f и используемые знания._


X - набор коротких сообщений  
x - конкретный короткий текст  
Y - метки трёх классов по языкам.  
y - вектор вероятностей принадлежности к классам  
$$f: X \to Y$$
f - это алгоритм, который по векторному представлению сообщения определяет язык на котором оно написано. 


Условие качества - минимизация ошибки определения языка. 
$$min~P(f(x) \ne y) \text{ на распределении } \mathcal{D}$$
  
Какие знания необходимы:  
* Алфавиты языков
* Словари языков для составления токенов
* Специфические символы, характерные для разных языков
